### Plotting Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tipsData = sns.load_dataset('tips')
```

```python
tipsData.shape
```

```python
tipsData[tipsData['total_bill'] > 20]
```

```python
x = tipsData['total_bill']
y = tipsData['tip']

plt.scatter(x, y, marker='.')
plt.xlabel('$ total_bill')
plt.ylabel('$ tip')
plt.title('Tips data')
plt.show()
```

```python
tipsData.smoker.value_counts()
```

```python
tipsData.day.value_counts()
```

```python
# autopct: auto percentage to show 2 decimal places 
plt.pie(tipsData.day.value_counts(), autopct='%.2f%%', labels=['Sat', 'Sun', 'Thurs', 'Fri'], explode=[0,0,0,0.2])
plt.show()
```

```python
plt.hist(tipsData['tip'], bins=10)
plt.title('Tip distribution')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.show()
```

```python
# basically if you have one column with multiple string values and you wanna plot them relative to another value then you can use stripplot
sns.stripplot(y='tip', x="time", data=tipsData, hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.pairplot(tipsData)
plt.show()
```
